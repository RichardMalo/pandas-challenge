```python
import pandas as pd
import os
import csv
```

```python
os.path.join("Resources", "budget_data.csv")
schools = os.path.join("Resources", "schools_complete.csv")
students = os.path.join("Resources", "students_complete.csv")
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(schools)
student_df = pd.read_csv(students)

# Combine the data into a single dataset.  
school_data_df = pd.merge(student_df, school_df, on=["school_name", "school_name"], how="left")
school_data_df.sample(5)
```

```python
school_df.sample(5)
```

```python
student_df.sample(5)
```

```python
# Total Number of Unique Schools
school_count = len(school_data_df["school_name"].unique())
print(school_count)
```

```python
# Total Students
student_count = len(school_data_df["Student ID"].unique())
print(student_count)
```

```python
# Total Budget
total_budget = school_df["budget"].sum()
total_budget
```

```python
# Average Math Score
average_math_score = school_data_df["math_score"].mean()
print(average_math_score)
```

```python
# Average Reading Score
average_reading_score = school_data_df["reading_score"].mean()
print(average_reading_score)
```

```python
# % passing math (the percentage of students who passed math)
passing_math_count = school_data_df[(school_data_df["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
print(passing_math_percentage)
```

```python
# % passing math (the percentage of students who passed math)
passing_read_count = school_data_df[(school_data_df["reading_score"] >= 70)].count()["student_name"]
passing_read_percentage = passing_read_count / float(student_count) * 100
print(passing_read_percentage)
```

```python
# % overall passing (the percentage of students who passed math AND reading)
passing_math_reading_count = school_data_df[(school_data_df["math_score"] >= 70) & (school_data_df["reading_score"] >= 70)].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
print(overall_passing_rate)
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = [{"Total Schools":school_count,
    "Total Students":student_count,
    "Total Budget":total_budget,
    "Average Math Score":average_math_score,
    "Average Reading Score":average_reading_score,
    "% Passing Math Rate":passing_math_percentage,
    "% Passing Reading Rate":passing_read_percentage,
    "% Overall Passing Rate":overall_passing_rate}]
# Create the Data Frame
district_summary_df = pd.DataFrame(district_summary)
# Format the Data Frame
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = (district_summary_df["Average Math Score"]).map("{:.2f}%".format)
district_summary_df["Average Reading Score"] = (district_summary_df["Average Reading Score"]).map("{:.2f}%".format)
district_summary_df["% Passing Math Rate"] = (district_summary_df["% Passing Math Rate"]).map("{:.2f}%".format)
district_summary_df["% Passing Reading Rate"] = (district_summary_df["% Passing Reading Rate"]).map("{:.2f}%".format)
district_summary_df["% Overall Passing Rate"] = (district_summary_df["% Overall Passing Rate"]).map("{:.2f}%".format)
#Print out the Data Frame
district_summary_df
```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
pd.cut(per_school_capita, spending_bins)
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = district_summary_perschool_df.groupby(["Spending Ranges (Per Student)"]).mean()["Per School Math Score:"]
spending_reading_scores = district_summary_perschool_df.groupby(["Spending Ranges (Per Student)"]).mean()["Per School Math Score:"]
spending_passing_math = district_summary_perschool_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math:"]
spending_passing_reading = district_summary_perschool_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading:"]
overall_passing_spending = district_summary_perschool_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing:"]
# Show the output
overall_passing_spending
```

```python
# Assemble into DataFrame
spending_summary_df = pd.DataFrame({"Average Math Score":spending_math_scores,
    "Average Reading Score":spending_reading_scores,
    "% Passing Math":spending_passing_math,
    "% Passing Reading":spending_passing_reading,
    "% Overall Passing":overall_passing_spending})

# Display Results
spending_summary_df
```

```python
# Apply formating to above to make decimals to a more traditional 2 digits
spending_summary_df["Average Math Score"]=spending_summary_df["Average Math Score"].map("{:.2f}%".format)
spending_summary_df["Average Reading Score"]=spending_summary_df["Average Reading Score"].map("{:.2f}%".format)
spending_summary_df["% Passing Math"]=spending_summary_df["% Passing Math"].map("{:.2f}%".format)
spending_summary_df["% Passing Reading"]=spending_summary_df["% Passing Reading"].map("{:.2f}%".format)
spending_summary_df["% Overall Passing"]=spending_summary_df["% Overall Passing"].map("{:.2f}%".format)
# Display Results
spending_summary_df
```
